# Race for iron: shaping tripartite community

This model was being developed collaboratively within the SFB MibiNet community to study how the structure of a synthetic microbial community is shaped by different siderophores

```python
import matplotlib.lines as mlines
import numpy as np
import pandas as pd
from matplotlib import colormaps
from matplotlib import pyplot as plt
from mxlpy import Simulator, plot, scan

from mxlmodels import get_dynamic_enterobactin
```

```python
_ = (
    Simulator(get_dynamic_enterobactin())
    .simulate(20)
    .get_result()
    .unwrap_or_err()
    .variables.plot()
)
```

```python
total_biomass = 0.11  # g/L

res = scan.time_course(
    get_dynamic_enterobactin(),
    to_scan=pd.DataFrame(
        [
            (f * total_biomass, (1 - f) * total_biomass)
            for f in np.linspace(0.05, 0.95, 19)
        ],
        columns=["x1", "x2"],
    ),
    time_points=np.linspace(0, 20, 101),
)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
res.variables["x1"].unstack().T.plot(
    ax=ax, lw="0.8", alpha=0.3, color="green", legend=False
)
res.variables["x2"].unstack().T.plot(
    ax=ax, lw="0.8", alpha=0.5, color="red", legend=False
)
res.variables["s1"].unstack().T.plot(
    ax=ax, lw="0.8", alpha=0.5, color="black", legend=False
)
plot.add_grid(ax)
ax.legend(
    handles=[
        mlines.Line2D([], [], color="green", label="x1"),
        mlines.Line2D([], [], color="red", label="x2"),
        mlines.Line2D([], [], color="black", label="s1"),
    ]
)
ax.set(
    title="",
    xlabel="time / h",
    ylabel="conc / (g / L)",
)
plt.show()
```

```python
cmap = colormaps["RdYlGn_r"]
colors = cmap(res.to_scan["x2"] / res.to_scan.sum(axis=1))

fig, (ax1, ax2) = plt.subplots(2, 1)

ax = res.fluxes["mu1"].unstack().T.plot(ax=ax1, color=colors, legend=False)
ax.set(
    ylim=(0.0, 0.5),
    xlabel="time / h",
    ylabel="growth rate e. coli / (1/h)",
)
plot.add_grid(ax)

ax = res.fluxes["mu2"].unstack().T.plot(ax=ax2, color=colors, legend=False)
ax.set(
    ylim=(0.0, 0.5),
    xlabel="time / h",
    ylabel="growth rate c. glut / (1/h)",
)
plot.add_grid(ax)
plt.show()
```
